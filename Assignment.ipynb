## Data Generation 

```python
import pandas as pd
import numpy as np
from faker import Faker
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fake=Faker()
```

```python
Faker.seed(35)
```

```python
np.random.seed(35)
```

```python
num_records=10000
```

```python
cities=['Ahmedabad','Surat','Mumbai','Pune','Bangalore','Hyderabad','Indore','Delhi']
```

```python
universities=['Gujarat Technological University','IIT','Delhi Unversity','Jain University','Manipal University', 'Jawaharlal Nehru Unversity','Amity University','Mumbai University']
```

```python
year=['1st','2nd','3rd','4th']
```

```python
programs=['Data Science','Robotics','AI','Electric vehicle','Machine Learning','Java','Autocad']
```

```python
sources=['Instagram','LinkedIn','College Collaboration','Google Form','Mass-Mailing','Whatsapp']
```

```python
data={
    'Lead_Id':[f"LD{i}" for i in range(1,num_records+1)],
    'Name':[fake.name() for _ in range(num_records)],
    'Gender':np.random.choice(['Male','Female'],size=num_records),
    'City':np.random.choice(cities,size=num_records),
    'University':np.random.choice(universities,size=num_records),
    'Year_Of_Study':np.random.choice(year,size=num_records),
    'Program_Interested':np.random.choice(programs,size=num_records),
    'Source':np.random.choice(sources,size=num_records)
}
```

## Budget Allocation 

```python
#compute the total leads that are converted to customers and total leads that are generated.
converted_yes_dict={}
total_lead_dict={}
for i in data['Source'].unique():
    converted_yes = data[(data['Source']==i) & (data['Converted']=='Yes')]['Converted'].value_counts().iloc[0]
    total_lead=data[(data['Source']=='College Collaboration')]['Source'].value_counts().iloc[0]
    converted_yes_dict[i]=converted_yes
    total_lead_dict[i]=total_lead
```

```python
converted_yes_dict
```

```python
total_lead_dict
```

```python
#compute the conversion rate of each source
conversion_rate={}
for i in sources:
    conversion_rate[i]=round(converted_yes_dict[i]/total_lead_dict[i]*100,2)
    
conversion_rate
```

```python
#compute the contribution of each source
contribution={}
Total_Yes_Leads=sum(converted_yes_dict.values())
for i in sources:
    contribution[i]=round(converted_yes_dict[i]/Total_Yes_Leads*100,2)
contribution
```

```python
#Budget Allocation for each source
budget_allocation={}
sum_of_conversion=sum(conversion_rate.values())
sum_of_contribution=sum(contribution.values())
#let's Total_Budget=100000
Total_Budget=100000
for i in sources:
    budget_allocation[i]=round(Total_Budget*((0.5*(conversion_rate[i]/sum_of_conversion))
                                             +(0.5*(contribution[i]/sum_of_contribution))),2)
    
budget_allocation
```

- **Most Of the budget should be allocated on Whatsapp, Google Form and Linkedin for marketing and targeting the potential customer.**
